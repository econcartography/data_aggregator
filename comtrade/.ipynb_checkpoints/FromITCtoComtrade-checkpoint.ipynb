```python
import pandas as pd
import os
import re
from functools import reduce
%matplotlib inline
```

```python
data_folder = '../ITCdata/Export_EU/'
data_processed_folder = 'data/'
```

```python
file_name = 'Trade_Map_-_List_of_products_exported_by_{country}.txt'
countries = [re.findall('by_(.+)\.txt', f)[0] for f in os.listdir(data_folder)]
print(countries)
```

```python
dfs = {c:pd.read_csv(data_folder+file_name.format(country=c), sep='\t') for c in countries}
```

```python
dfs['Latvia'].head()
```

```python
# Removing 'Unnamed: 20'
[df.drop(columns='Unnamed: 20', inplace=True) for df in dfs.values() if 'Unnamed: 20' in df.columns ];
# renaming the columns of export with the only year
[df.rename(columns={c:'Exported' in c and c.split(' ')[-1] or c for c in df.columns}, inplace=True) for df in dfs.values()];
[df.rename(columns={'Product label':'Product_label'}, inplace=True) for df in dfs.values()];
for c in countries:
    # adding a column for the country
    dfs[c]['Country']=c
    # removing the total value
    if len(dfs[c][dfs[c].Code=='TOTAL']): dfs[c].drop([0], inplace=True)
```

```python
dfs['Romania'].head()
```

```python
for c in countries:
    df_y = {}
    for y in range(2001,2019):
        df_y[y] = dfs[c][['{}'.format(y),'Country','Code','Product_label']].copy()
        df_y[y].rename(columns={'{}'.format(y):'Export'}, inplace=True)
        df_y[y]['Year'] = y
        df_y[y] = df_y[y][['Year','Country','Code','Product_label','Export']]
    dfs[c] = reduce(lambda df0,df1: df0.append(df1), df_y.values())
```

```python
dfs['Romania'].head()
```

```python
df_export = reduce(lambda df0,df1: df0.append(df1, ignore_index=True), dfs.values())
```

```python
df_export.head()
```

```python
df_yc_exportTot = df_export.groupby(['Year', 'Country']).sum()
df_yc_exportTot.head()
```

```python
df_export = pd.merge(df_export, df_yc_exportTot, on=['Year','Country'], suffixes=('','_tot_country'))
df_export.head()
```

```python
df_yp_exportByProd = df_export.groupby(['Year', 'Code']).sum()
df_yp_exportByProd.head()
```

```python
df_export = pd.merge(df_export, df_yp_exportByProd, on=['Year','Code'], suffixes=('','_tot_product'))
```

```python
df_export.head()
```

```python
df_exportByYear = df_export.groupby(['Year']).Export.sum()
df_exportByYear.head()
```

```python
df_export = pd.merge(df_export, df_exportByYear, on='Year', suffixes=('','_tot_year'))
```

```python
df_export.head()
```

```python
gb = df_export.groupby(['Year','Country','Code']).Export.sum().unstack(2)
gb.head(20)
```

```python
for y in range(2001,2019):
    gb.loc[y].to_csv(data_processed_folder+'export_matrix_{}.csv'.format(y), index=True)
```

RCA filter:

$$\frac{e_{c,p}}{\sum_p e_{c,p'}} > \frac{\sum_{c'} e_{c,p'}}{\sum_{c',p'} e_{c',p'}}$$

```python
df_export['RCA'] = (df_export.Export/df_export.Export_tot_product) / (df_export.Export_tot_country/df_export.Export_tot_year)
```
